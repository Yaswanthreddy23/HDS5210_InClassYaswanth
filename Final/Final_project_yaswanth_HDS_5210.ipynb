# ***Data Merging***

# Merging 1

This step involves merging the patient data with the hospital visit data to create a unified dataset for analysis. The merging operation uses the ID column from the patient data and the Patient_ID column from the hospital data as the linking key. After successfully merging, the redundant Patient_ID column is removed to streamline the resulting DataFrame.

This merged dataset provides a consolidated view of patient demographics and hospital visit details, enabling easier access to combined information for further analysis, such as understanding patient outcomes, hospital resource utilization, or patient-hospital interactions.

```python
# Merging 1

def join_patient_hospital_data(patient_df, hospital_df):

    """
    Joins the patient data with the hospital visit data based on the ID and Patient_ID.

    Arguments:
    patient_df: DataFrame containing patient details (ID, Name, Age, etc.)
    hospital_df: DataFrame containing hospital visit details (Patient_ID, Hospital, Doctor, etc.)

    Returns: Merged DataFrame containing patient and hospital details.
    """

    # Merging patient data with hospital data on patient ID
    merged_df = pd.merge(patient_df, hospital_df, left_on='ID', right_on='Patient_ID', how='inner')

    # Dropping the redundant 'Patient_ID' column as it is already represented by 'ID'
    merged_df.drop(columns=['Patient_ID'], inplace=True)

    return merged_df


# Join patient_df with hospital_df
merged_patient_hospital_df = join_patient_hospital_data(patient_df, hospital_df)
merged_patient_hospital_df.head()
```

# ***Data Aggregation and Pivoting***

# Aggregation 1

This step involves the aggregation of total billing amounts by grouping the data based on the Medical Condition field. The function uses the groupby method to calculate the sum of the Billing Amount for each medical condition and then sorts the results in descending order to highlight conditions with the highest billing totals.

This aggregation simplifies the analysis of billing trends, enabling a clear understanding of which medical conditions incur the highest costs. It provides valuable insights for hospital management and healthcare policymakers to allocate resources effectively and identify cost drivers in patient care.

```python
# Aggregation 1
def aggregate_billing_by_condition(df):

    """
    Aggregates the total billing amount by medical condition.

    Returns the DataFrame with total billing amounts aggregated by medical condition.
    """

    # Aggregating total billing amount by 'Medical Condition'
    billing_by_condition = df.groupby('Medical Condition')['Billing Amount'].sum().reset_index()

    # Sorting the result to have the highest billing condition at the top
    billing_by_condition = billing_by_condition.sort_values(by='Billing Amount', ascending=False)

    return billing_by_condition



# Aggregation 1: Total Billing Amount by Medical Condition
billing_by_condition = aggregate_billing_by_condition(final_df)
print("Total Billing Amount by Medical Condition:\n", billing_by_condition)
```

# Aggregation 2

This step involves aggregating the average age of patients by grouping the data based on the Hospital and Admission Type fields. The function calculates the mean age for each combination of hospital and admission type, providing insights into the demographics of patients served under different admission circumstances.

The results are sorted in descending order of age for better readability, highlighting hospitals and admission types with older patient populations. This aggregation helps in understanding the demographic trends across hospitals and admission types, enabling healthcare providers to tailor services and resources to the specific needs of different age groups.

```python
# Aggregation 2
def aggregate_age_by_hospital_admission(df):
    """
    Aggregates the average age of patients by hospital and admission type.

    Returns the DataFrame with average age aggregated by hospital and admission type.
    """

    # Aggregating average age by 'Hospital' and 'Admission Type'
    avg_age_by_hospital_admission = df.groupby(['Hospital', 'Admission Type'])['Age'].mean().reset_index()

    # Sorting the result for better readability
    avg_age_by_hospital_admission = avg_age_by_hospital_admission.sort_values(by='Age', ascending=False)

    return avg_age_by_hospital_admission



# Aggregation 2: Average Age by Hospital and Admission Type
avg_age_by_hospital_admission = aggregate_age_by_hospital_admission(final_df)
print("Average Age by Hospital and Admission Type:\n", avg_age_by_hospital_admission)
```

# Pivoting 1

This step involves pivoting the billing data to create a matrix where each row represents a hospital, and each column represents a medical condition. The values in the matrix correspond to the total billing amount for that condition at each hospital. The pivot_table function is used to achieve this, with missing values filled as 0 for better completeness.

This pivoted structure allows for an intuitive comparison of billing amounts across different hospitals and medical conditions. It simplifies the analysis by transforming the data into a format where patterns and disparities in billing practices by hospital and condition can be easily identified, enabling targeted resource allocation and cost management.

```python
# pivoting 1
def pivot_billing_by_condition(df):

    """
    Pivots the billing information so that each medical condition has its own column for each hospital.

    Returns the Pivoted DataFrame with hospitals as rows and medical conditions as columns.
    """

    # Pivoting the data to show total billing amount for each condition by hospital
    pivoted_billing = df.pivot_table(values='Billing Amount',
                                     index='Hospital',
                                     columns='Medical Condition',
                                     aggfunc='sum',
                                     fill_value=0)

    return pivoted_billing


# Pivoting 1: Billing by Condition (Pivoting hospitals and conditions)
pivoted_billing = pivot_billing_by_condition(final_df)
print("Pivoted Billing by Condition:\n", pivoted_billing)
```

# Transformation 3

This step involves creating a new column called **'Age Group'**, which categorizes patients into groups based on their age: **'Young' (≤18 years)**, **'Adult' (19–60 years)**, and **'Senior' (>60 years)**. The transformation uses a helper function to assign these categories to each patient based on the value in the `Age` column.

This transformation simplifies the analysis by grouping patients into meaningful age categories. It enables insights into age-related trends, such as medical conditions, hospital stays, or billing amounts, while reducing the complexity of working with continuous age values. This categorization is particularly useful for demographic analysis and reporting.

```python
# Transformation 3

def create_age_group(df):

    """
    Creates a new 'Age Group' column based on the 'Age' column.

    Returns DataFrame with 'Age Group' column added.
    """
    def categorize_age(age):
        if age <= 18:
            return 'Young'
        elif 19 <= age <= 60:
            return 'Adult'
        else:
            return 'Senior'

    df['Age Group'] = df['Age'].apply(categorize_age)
    return df


final_df = create_age_group(final_df)

# View the transformed data
final_df.head()
```

# Transformation 4

This step involves **standardizing the 'Medical Condition' column** by capitalizing the first letter of each value. The transformation is performed using a lambda function that applies the `capitalize()` method to each entry in the column.

This transformation ensures uniform formatting of the `Medical Condition` field, enhancing data quality and readability. It eliminates inconsistencies in case formatting, which can be critical for accurate grouping, filtering, and analysis based on medical conditions. This step is particularly useful for creating polished reports and ensuring consistency in downstream analysis.

```python
# Transformation 4

def standardize_medical_condition(df):
    """
    Standardizes the 'Medical Condition' field by capitalizing the first letter.

    Returns DataFrame with standardized 'Medical Condition' column.
    """

    df['Medical Condition'] = df['Medical Condition'].apply(lambda x: x.capitalize())

    return df

final_df = standardize_medical_condition(final_df)

# View the transformed data
final_df.head()
```

# Transformation 5

This step involves calculating the **duration of hospital stay** for each patient in days by subtracting the `Date of Admission` from the `Discharge Date`. The resulting values are stored in a new column, **'Duration of Stay'**, which provides a clear measure of the length of each patient's hospital stay.

This transformation adds a critical metric to the dataset, enabling the analysis of stay durations across different groups, such as age, medical condition, or hospital. It supports insights into resource utilization, patient care efficiency, and trends in hospitalization durations, which are valuable for hospital management and operational planning.

```python
# Transformation 5

def calculate_duration_of_stay(df):

    """
    Calculates the duration of stay for each patient (in days) based on the 'Date of Admission' and 'Discharge Date'.

    Returns DataFrame with new 'Duration of Stay' column added.
    """

    df['Duration of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

    return df

final_df = calculate_duration_of_stay(final_df)

# View the transformed data
final_df.head()
```
